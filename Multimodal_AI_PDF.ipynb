```python
from google.colab import auth
auth.authenticate_user()

from rich.console import Console
console = Console()
```

```python
from google.colab import userdata
PROJECT_ID = userdata.get('PROJECT_ID')

import vertexai
vertexai.init(project=PROJECT_ID, location="us-central1")
```

```python
from vertexai.generative_models import GenerativeModel, Part

model = GenerativeModel("gemini-1.5-flash-001")

prompt = """
You are an expert evaluator. You have the ability to understand an input document and generate questions based on its content.
Generate a structured JSON object that contains 10 questions based on the entire document. Complex questions for each chapter or section. Every question should have 5 options, only one option should be the correct.
Generate all the questions based on the document context ONLY.
"""

pdf_file = Part.from_uri("gs://questionsanswersproject/software-architecture-for-developers.pdf", mime_type="application/pdf")
response = model.generate_content([pdf_file, prompt])
console.print(response.text)

```
